```python
from ultralytics import YOLO
```

```python
LAST_MODEL='/content/drive/MyDrive/ball_detection/models/80/train/weights/last.pt'
BEST_MODEL='/content/drive/MyDrive/ball_detection/models/80/train/weights/best.pt'
DATA='/content/drive/MyDrive/ball_detection/datasets/ball/ball.yaml'
PROJECT='/content/drive/MyDrive/ball_detection/models/90'
```

```python
MB_PROJECT='/content/drive/MyDrive/ball_detection/models/70_and_10mb'
MB_DATA='/content/drive/MyDrive/ball_detection/datasets/ball_copy/ball.yaml'
```

```python
model = YOLO(LAST_MODEL)
```

```python
# !rm -rf /content/drive/MyDrive/ball_detection/models/70_and_10mb
```

```python
model.train(
    data=DATA,
    epochs=10,
    warmup_epochs=0.5,

    lr0=0.003,
    optimizer="AdamW",
    cos_lr=True,

    imgsz=1600,
    batch=4,

    project=PROJECT,

    cls=0.5,
    box=8.0,
    dfl=1.8,

    scale=0.3,
    mixup=0.0,

    degrees=0,
    translate=0.10,
    flipud=0.0,
    fliplr=0.5,

    patience=0,
)
```

```python
!rm -rf /content/drive/MyDrive/ball_detection/models/80/val
```

```python
model.val(
    data=DATA,
    imgsz=1600,
    batch=16,
    name="val",
    project=PROJECT
)
```
