```python
from ultralytics import YOLO
```

```python
LAST_MODEL='/content/drive/MyDrive/player_detection/models/50/train/weights/last.pt'
BEST_MODEL='/content/drive/MyDrive/player_detection/models/50/train/weights/best.pt'
DATA='/content/drive/MyDrive/player_detection/datasets/player/player.yaml'
PROJECT='/content/drive/MyDrive/player_detection/models/80'
```

```python
model = YOLO(LAST_MODEL)
```

```python
model.train(
    data=DATA,
    epochs=30,
    warmup_epochs=1.0,

    lr0=0.0015,             # players â†’ more stable
    optimizer="AdamW",
    cos_lr=True,

    imgsz=640,
    batch=48,

    project=PROJECT,
    # patience=0,

    # --- LOSS WEIGHTS (players are medium objects) ---
    cls=0.8,                # stronger class separation
    box=6.0,                # slightly lower (players less tiny than balls)
    dfl=1.6,

    # --- AUGMENTATION (tailored for football players) ---
    scale=0.15,             # players get distorted at high scale
    mixup=0.0,              # BAD for player shapes â†’ keep off

    degrees=0,
    translate=0.05,         # keep small movement to avoid weird positioning
    flipud=0.0,
    fliplr=0.4,             # football â†’ moderate horizontal flips OK

    shear=0.0,
    perspective=0.0,
)
```

```python
model.val(
    data=DATA,
    imgsz=640,
    batch=2,
    project=PROJECT,
    name="val"
)
```
