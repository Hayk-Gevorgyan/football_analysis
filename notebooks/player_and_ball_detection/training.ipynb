```python
from ultralytics import YOLO
```

# State paths

```python
MODEL_PATH = '/content/drive/MyDrive/football_analysis/training/birds_eye_view_yolov12m/runs/30/weights/last.pt'
PROJECT = '/content/drive/MyDrive/football_analysis/training/birds_eye_view_yolov12m/runs'
DATA = '/content/drive/MyDrive/football_analysis/training/birds-eye-view/data.yaml'
```

# Initialize the model

```python
model = YOLO(MODEL_PATH)
```

# Train the loaded model with the new dataset

```python
model.train(
    data=DATA,
    epochs=100,
    warmup_epochs=2,
    lr0=0.005,
    imgsz=1088,
    batch=2,
    project=PROJECT,
    name='35',
    cls=1.0,
    box=12.0,
    dfl=2.5,
    scale=0.1,
    mixup=0.15,
    degrees=0,
    translate=0.05,
    flipud=0.0,
    fliplr=0.5,
)
```

## Validate custom model
